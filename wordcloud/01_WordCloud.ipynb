```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Alice in Wonderland

```python
with open('data/Alice.txt','r') as f:
    text = f.read()
stopwords = set(STOPWORDS)
len(stopwords)  #불용어
```

```python
#stopwords에 said 추가
stopwords.add('said')
```

```python
wc = WordCloud(background_color='white', max_words=2000, stopwords=stopwords)
wc = wc.generate(text)
#wc = 워드클라우드
```

```python
keys = list(wc.words_.keys())
values = list(wc.words_.values())
keys[0]

for i in range(10):
    print(f'{keys[i]} : {values[i]}')
```

```python
fig = plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
with open('/data/A_new_hope.txt','r') as f:
    text = f.read()
text= text.replace('HAN','Han')
text= text.replace("LUKE's",'Luke')
mask= np.array(Image.open('/Users/daeun/Desktop/Workspace/DataAnalysis/07.NLP/data/Stormtrooper_mask.png'))
```

```python
stopwords= set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')
```

```python
wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, margin=10, random_state=1).generate(text)
default_colors = wc.to_array()
```

```python
import random
def grey_color(word, font_size, position, orientation, random_state=None, **kwargs):
    return 'hsl(0, 00%, %d%%)' % random.randint(60,100)
    #      hue saturation lightness
    return f'hsl(0, 0%, {random.randint(60,100)}%)'
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```
