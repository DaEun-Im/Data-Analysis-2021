### 앤스콤 데이터로 연습하기

```python
%matplotlib inline
ans = sns.load_dataset('anscombe')
ans.head()
```

```python
dataset_1 = ans[  ans['dataset'] == 'I' ]
dataset_1
```

```python
dataset_2 = ans[  ans['dataset'] == 'II' ]
dataset_3 = ans[  ans['dataset'] == 'III' ]
dataset_4 = ans[  ans['dataset'] == 'IV' ]
```

```python
#그래프 격자가 위치할 기본 틀 만들기
fig = plt.figure()    #가장 커다란 틀

#subplot 만들기
axes1 = fig.add_subplot(2,2,1)  #add_subplo(행의수,열의수,위치)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
```

```python
axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

fig
```

```python
#각각의 그래프가 무엇인지 쉽게 구분할 수 있도록 각 그래프에 제목 붙이기

axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')

fig
```

```python
#기본 틀(fig, 가장 커다란 틀) 에 제목 붙이기
fig.suptitle('Anscombe Data')
#   super_title
fig
```

```python
fig.tight_layout()
fig
```

### matplotlib 라이브러리 자유자재로 사용하기

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#기본틀 마련
fig = plt.figure()

#격자 구성
axes1 = fig.add_subplot(1,1,1)
```

```python
# 히스토그램 생성하기 (히스토그램: 변수의 분포, 빈도)
axes1.hist(tips['total_bill'], bins=10)  #bins = x축 간격
axes1.set_title('Histogram of Total Bill')  #set_title('') 그래프 제목
axes1.set_xlabel('Frequency') #set_xlabel('') x축제목
axes1.set_ylabel('Total Bill') #set_ylabel('') y축제목

fig
```

```python
# 산점도그래프 (Scatter Plot) 그리기 - 변수 2개사용(이변량 그래프)
scatter_plot = plt.figure()  #기본틀생성

axes1 = scatter_plot.add_subplot(1,1,1) #subplot(axes)생성

axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('ScatterPlot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# boxplot(박스그래프) 그리기 - 이산형변수(명확히구분되는값), 연속형변수(명확히셀수없는값) 함께 사용

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

```
