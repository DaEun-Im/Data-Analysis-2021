### 1. Bar Chart 
-  x 데이터가 카테고리값인 경우

```python
y = [2, 3, 1]
x = np.arange(len(y))  #array([0, 1, 2])
xlabel = ['가', '나', '다']

plt.title("Bar Chart")
plt.bar(x, y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
np.random.seed(0)

people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

### 2. Pie Chart 

- 카테고리 별 값의 상대적인 비교를 해야 할 때

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### 3. histogram

- hist 명령은 bins 인수로 데이터를 집계할 구간 정보를 받는다. 반환값으로 데이터 집계 결과를 반환한다.

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.title("Histogram")
arrays, bins, patches = plt.hist(x, bins=10)
plt.show()
```

### 4. 산점도 (Scatter Plot)

- 두 개의 실수 데이터 집합의 상관관계를 살펴볼 때 사용

```python
np.random.seed(0)
X = np.random.randn(100)
Y = np.random.randn(100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```

```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3)  #c=color, s=size
plt.show()
```

- Anscombe's Quartet

```python
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.tail()
```

```python
ans.groupby('dataset').describe()
# dataset1,2,3,4 각각 11개씩의 데이터
```

```python
ans[ans.dataset == 'I']
```

```python
fig, axes= plt.subplots(2,2, figsize=(10,8),
                        sharex=True, sharey=True)

for index, roman in enumerate(['I','II','III','IV']):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.plot(ds.x, ds.y, 'o', markersize=10)
    ax.plot( [3,20],[4.5,13], 'r-', lw=2 )
    ax.set_title(f'Dataset {roman}')

fig.suptitle("Anscombe's Quartet", fontsize=16)
plt.tight_layout()
plt.show()
```
