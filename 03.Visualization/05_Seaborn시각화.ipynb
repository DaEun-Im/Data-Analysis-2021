```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
mpl.rcParams['axes.unicode_minus'] = False
font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf'
plt.rc('font', family='AppleGothic')
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터
tips = sns.load_dataset("tips")    # 팁 데이터
flights = sns.load_dataset("flights")    # 여객운송 데이터
```

- countplot

1. 각 카테고리 값별로 데이터가 얼마나 있는지 표시할 수 있다

2. 데이터프레임에서만 사용 가능


3. sns.countplot(x="column_name", data=dataframe)

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉호의 각 클래스별 승객 수')
plt.show()
```

```python
sns.countplot(x="day", data=tips)
plt.title("요일별 팁을 준 횟수")
plt.show()
```

- pairplot  (다차원 실수형 데이터)

1. 만약 3차원 이상의 데이터라면 seaborn 패키지의 pairplot 명령을 사용

2. pairplot은 "데이터프레임을 인수로 받아" 그리드(grid) 형태로 "각 데이터 열의 조합에 대해 스캐터 플롯"을 그린다.

3. 같은 데이터가 만나는 대각선 영역에는 해당 데이터의 히스토그램을 그린다.

```python
sns.pairplot(iris)
plt.title('Iris 데이터의 Pair Plot')
plt.show()
```

```python
sns.pairplot(iris, hue="species", markers=)
plt.title("Iris Pair Plot, Hue로 꽃의 종을 시각화")
plt.show()
- #카테고리형 데이터가 섞여 있는 경우에는 hue 인수에 카테고리 변수 이름을 지정하여 카테고리 값에 따라 색상을 다르게 할 수 있다.
```

- heatmap

```python
flights_passengers = flights.pivot('month','year','passengers')
plt.figure(figsize=(8,6))
sns.heatmap(flights_passengers, annot=True, fmt='d', linewidths=1)
plt.show()
```

- barplot

: 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 기본적인 바 차트를 생성한다.

평균은 막대의 높이로, 편차는 에러바(error bar)로 표시한다.

```python
sns.boxplot(x='day',y='total_bill', hue='sex', data=tips)
plt.title('요일 별, 전체 요금')
plt.show()
```

- boxplot

```python
sns.boxplot(x='day',y='total_bill', hue='sex', data=tips)
plt.title('요일 별, 전체 요금의 BoxPlot')
plt.show()
```

```python
#hue='sex'
#두 가지 카테고리 값에 의한 실수 값의 변화를 보기 위한 hue 인수를 제공한다.
#hue 인수에 카테고리 값을 가지는 변수의 이름을 지정하면 카테고리 값에 따라 다르게 시각화된다
sns.barplot(x="day", y="total_bill", hue='sex', data=tips)
plt.title("요일 별 , 성별 전체 요금")
plt.show()
```

```python
np.random.seed(0)
sns.stripplot(x="day", y="total_bill", hue="sex", data=tips, jitter=True)
plt.title("요일 별, 성별 전체 팁의 Strip Plot")
plt.legend(loc=1)
plt.show()
```
