```python
import numpy as np 
import pandas as pd 
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
mpl.rcParams['axes.unicode_minus'] = False
font_path = '/System/Library/Fonts/Supplemental/AppleGothic.ttf'
plt.rc('font', family='AppleGothic')
```

```python
np.random.seed(0)
df1 = pd.DataFrame(np.random.randn(100, 3),
                   index=pd.date_range('1/1/2021', periods=100),
                   columns=['A', 'B', 'C']).cumsum()
df1.tail()
```

```python
df1.plot()
plt.title("Pandas의 Plot메소드 사용 예")
plt.xlabel("시간")
plt.ylabel("Data")
plt.grid()
plt.show()
```

```python
iris = sns.load_dataset("iris")    # 붓꽃 데이터
titanic = sns.load_dataset("titanic")    # 타이타닉호 데이터

iris.sepal_length[:20].plot(kind='bar', rot=0)
plt.title("꽃받침의 길이 시각화")
plt.xlabel("Data")
plt.ylabel("꽃받침의 길이")
plt.show()
```

```python
iris[:5].plot.bar(rot=0)
plt.title("Iris 데이터의 Bar Plot")
plt.xlabel("Data")
plt.ylabel("각 Feature의 값")
plt.ylim(0, 7)
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'feature'
df2
```

```python
df2.plot.bar()
plt.title('각 종의 Feature별 평균')
plt.xlabel('평균')
plt.ylabel('단위(cm)')
plt.ylim(0,8)
plt.grid()
plt.show()
```

```python
df2.T.plot.bar()
plt.title('각 Feature의 품종별 평균')
plt.xlabel('평균')
plt.ylabel('단위(cm)')
plt.ylim(0,8)
plt.grid()
plt.show()
```

```python
df3 = titanic['class'].value_counts()
df3.plot.pie(autopct='%.2f%%')
plt.title("선실별 승객 수 비율")
plt.axis('equal')
plt.show()
```

```python
iris.plot.box()
plt.title("각 Feature 값들의 빈도수에 대한 Box Plot")
plt.xlabel("Feature")
plt.ylabel("데이터 값")
plt.show()
```

```python
iris.boxplot(by='species', figsize=(10,8))
plt.tight_layout(pad=1, h_pad=1)
plt.title("각 Feature의 종 별 데이터에 대한 Box Plot")
plt.show()
```
