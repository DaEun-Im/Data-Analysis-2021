### 1.라인플롯

```python
plt.plot([1,4,9,16]) # X=[0,1,2,3] 가정  # x값을 안줬으므로 임의로 주어짐-> range(n) , n=y값갯수
plt.show()
```

```python
plt.plot([1,2,3,4], [1,4,9,16])  #x값, y값 둘다 줌
plt.title('y=x**2 그래프')   #그래프 제목
plt.grid()  #격자무늬
plt.show()  #그래프출력
```

- 스타일지정

```python
# 색깔(color), 마커(market), 선 종류(line style)
# 'rs--'
# r=색깔(red), s=
# 격자점모양(square), --:점선
plt.plot([1,2,3,4], [1,4,9,16], 'rs--') 
plt.title(" 'rs--' 스타일 ")  #그래프 제목
plt.grid()  #격자무늬
plt.show()  #그래프출력
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro--') 
plt.title(" 'ro--' 스타일 ")  #그래프 제목
plt.grid()  #격자무늬
plt.show()  #그래프출력
```

```python
plt.plot([1,2,3,4], [1,4,9,16], c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r") 
plt.title(" x축, y축의 범위 설정 ")  #그래프 제목
plt.xlim(0,5)
plt.ylim(-2,20)
plt.grid()
plt.show()
```

- 틱 설정

```python
# 플롯이나 차트에서 축상의 위치 표시 지점을 틱(tick)이라고 하고
# 이 틱에 써진 숫자 혹은 글자를 틱 라벨(tick label)이라고 한다.
# 틱의 위치나 틱 라벨은 맷플롯리브가 자동으로 정해주지만
# 만약 수동으로 설정하고 싶다면 xticks 명령이나 yticks 명령을 사용한다.
```

```python
X = np.linspace(-np.pi, np.pi, 300)
Y = np.cos(X)

plt.plot(X, Y)
plt.grid()


plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 300)
Y = np.cos(X)

plt.plot(X, Y)
plt.grid()

plt.xticks( [-np.pi, -np.pi/2 , 0 , np.pi/2 , np.pi] )  #pi=3.14
plt.yticks( [-1 , 0  , 1] )

plt.title("x축과 y축의 tick label 설정")

plt.show()
```

- 여러개의 선 그리기

```python
t = np.arange(0., 5., 0.2)
plt.title=('라인 플롯에서 여러 개의 선 그리기')
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t , 0.2* t**3, 'g^-')
plt.grid()
plt.show()
```

- 범례(Legend)

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("legend를 표시한 플롯")
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.grid()
plt.show()
```

```python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.title("x축 y축 라벨을 표시한 플롯")
plt.plot(X, C, label="cosine")
plt.plot(X, S, label="sine")
plt.legend(loc=2)
plt.xlabel('x축라벨'); plt.ylabel('y축라벨')
plt.grid()
plt.show()
```
